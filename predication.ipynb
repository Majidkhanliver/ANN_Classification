```python
import pickle
import pandas as pd
```

```python
## load the Pickle files
with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder_gender = pickle.load(f)
with open('onehot_encoder_geography.pkl', 'rb') as f:       
    onehot_encoder_geography = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f) 
```

```python
## Load the trained model
from tensorflow.keras.models import load_model
model = load_model('model.h5')
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Female',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_data['Gender']=label_encoder_gender.transform([input_data['Gender']])[0]
```

```python
input_data
```

```python
onehot_encoder_geography.categories_  # To check categories for Geography
```

```python
geography_encoded = onehot_encoder_geography.transform([[input_data['Geography']]]).toarray()

geography_encoded_df = pd.DataFrame(geography_encoded, columns=onehot_encoder_geography.get_feature_names_out(['Geography']))
```

```python
geography_encoded_df
```

```python
input_data = pd.concat([pd.DataFrame([input_data]).drop('Geography', axis=1), geography_encoded_df], axis=1)
```

```python
input_data
```

```python
input_data
```

```python
## Scaling the input data
input_data_scaled = scaler.transform(input_data)
```

```python
input_data_scaled
```

```python
## Prediction
prediction = model.predict(input_data_scaled)
```
